## Insights :

## 🧩 Conclusion

The HR Analytics EDA provided a clear understanding of the key factors influencing employee attrition in the organization.

### Key Takeaways:
- Employees with **lower monthly income**, **fewer years at the company**, and **longer commute distances** showed a higher chance of leaving.
- **R&D and Sales departments** experienced more attrition compared to other departments.
- Employees who **haven’t been promoted in several years** or have **low job satisfaction** are more likely to quit.
- **Younger employees** and those in **entry-level roles** tend to change jobs more frequently.
- **Work-life balance** and **job involvement** also play a significant role in retention.

### Overall Summary:
The analysis highlights that **attrition is multi-factorial** — mainly influenced by compensation, career growth opportunities, and employee satisfaction.  
By focusing on **fair salary structures**, **timely promotions**, and **better work engagement**, HR teams can significantly reduce employee turnover and improve overall organizational health.
